Motivation:
Using the previous algorithm, values tended to stray away from 15fps expected values, hence, for larger videos it might be the case that more if statements would be required. 

Using the new algorithm, we cast a larger net as more frames are dropped. 

```python
import pandas as pd 
import numpy as np 
```

```python
file = "~/Lab/timeStamps.csv"
videofile = "~/Lab/0.avi" #filepath for video 

timestamps = pd.read_csv(file)
with open("/Users/nelsin/Lab/timeStamps.csv", "r") as file_for_rows:
    data = file_for_rows.readlines()

amount_of_rows_string = ''
last_line = data[-1]
index = 0

#Loop through the last line of the index searching for a comma (separator)
#Add all characters that appear before the comma to a string
while last_line[index] != ",":
    amount_of_rows_string = amount_of_rows_string + last_line[index] #goddamn python
    index = index + 1

#Error check: what if last row is empty?
amount_of_rows = int(amount_of_rows_string)

```

```python
offset = timestamps["Time Stamp (ms)"][0] #Offset = difference between first time stamp and 0.
timestamps["Shifted Times (ms)"] = timestamps["Time Stamp (ms)"] - offset #Adjust time stamps by offset


for x in range(amount_of_rows):   #check that this does not need to be increased by 1
    timestamps.at[x, "Expected Times (ms)"] = x* (1/0.015)


```

```python
from os import times
from time import time

added_frames = []

def readjust_lower_rows(start, end):
    for index2 in range(start+2, end):
        modified_entry = [timestamps.at[index2, "Frame Number"], timestamps.at[index2, "Time Stamp (ms)"], timestamps.at[index2, "Buffer Index"], timestamps.at[index2, "Shifted Times (ms)"], timestamps.at[index2, "Expected Times (ms)"] + (1/0.015)]
        timestamps.loc[index2] = modified_entry

two = 0
for index in range(amount_of_rows):
    difference = timestamps.at[index, "Shifted Times (ms)"] - timestamps.at[index, "Expected Times (ms)"]
    if (difference > (1/0.015)): #Consider a frame to be dropped
        added_frames.append(index + 1)
        shifted_time = timestamps.at[index, "Shifted Times (ms)"]
        expected_time = timestamps.at[index, "Expected Times (ms)"] + (1/0.015)
        new_entry = [index + 1, 0, 1.0, shifted_time, expected_time]
        column_names = ['Frame Number', 'Time Stamp (ms)', 'Buffer Index', 'Shifted Times (ms)', 'Expected Times (ms)'] 
        timestamps = pd.DataFrame(np.insert(timestamps.values, index+1, new_entry, axis=0), columns=column_names)
        readjust_lower_rows(index, amount_of_rows)
        index=index+2
        

     


pd.set_option('display.max_rows', None)
timestamps
```
